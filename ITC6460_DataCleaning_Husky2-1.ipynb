```python
merged_df = pd.merge(left=df1, right=df2, on=['Name','Platform', 'Publisher', 'Year'], how='inner')
```

```python
merged_df.isnull().mean() * 100
```

```python
merged_df = merged_df.drop(columns=['Rank_x','Rank_y'])
```

```python
merged_df.isnull().mean() * 100
```

```python
merged_df
```

```python
from sklearn.impute import KNNImputer

# Initialize the KNN Imputer
knn_imputer = KNNImputer(n_neighbors=5)

# It's important to scale your features before applying KNN imputation
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

# Assuming 'df' is your DataFrame and 'Critic_Score' is among other numerical columns you wish to include
numerical_features = merged_df.select_dtypes(include=['float64', 'int64']).columns.tolist()
df_scaled = scaler.fit_transform(merged_df[numerical_features])
df_scaled = pd.DataFrame(df_scaled, columns=numerical_features)

# Apply KNN imputation
df_scaled_imputed = knn_imputer.fit_transform(df_scaled)
df_scaled_imputed = pd.DataFrame(df_scaled_imputed, columns=numerical_features)

# Inverse transform to get back to the original scale
df_imputed = scaler.inverse_transform(df_scaled_imputed)
df_imputed = pd.DataFrame(df_imputed, columns=numerical_features)

# Update the original DataFrame with the imputed values
merged_df['Critic_Score'] = df_imputed['Critic_Score']
```

```python
merged_df.isnull().mean() * 100
```

```python
merged_df
```

```python
# Replace NaN values in 'Year' column with a placeholder and convert to integer
merged_df['Year'] = merged_df['Year'].fillna(-1).astype(int)
merged_df['Year_as_date'] = pd.to_datetime(merged_df['Year'], format='%Y', errors='coerce')
```

```python
# We'll describe the sales columns to see the distribution and identify potential outliers
merged_df[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']].describe()
```

```python
# Step 4: Normalize text data (convert to consistent casing)
merged_df['Name'] = merged_df['Name'].str.title()
merged_df['Platform'] = merged_df['Platform'].str.upper()
merged_df['Genre'] = merged_df['Genre'].str.title()
merged_df['Publisher'] = merged_df['Publisher'].str.title()
```

```python
merged_df.to_csv('p1.csv', index=False)
```
